```python
def get_recipe_info(recipe_title):
    recipe_dic = [i for i in recipe_full if "title" in i.keys() and i["title"]==recipe_title]
    if len(recipe_dic) == 0:
        return pd.Series({"len_directions":None, "len_ingredients":None})
    else:
        recipe_dic = recipe_dic[0]
    len_directions = len(recipe_dic["directions"])
    len_ingredients = len(recipe_dic["ingredients"])
    return pd.Series({"len_directions":len_directions, "len_ingredients":len_ingredients})
```

```python
data_2[["len_directions","len_ingredients"]] = data_2.title.parallel_apply(lambda x : get_recipe_info(x))
```

Clustering 

```python
from sklearn import datasets 
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
recipes_2 = recipes.copy()
```

```python
recipes_2 = recipes_2.drop(columns =['directions', 'date', 'categories', 'title', 'desc', 'ingredients'])
display(recipes_2)
```

```python
import pickle

scaler =StandardScaler()
scaler.fit(recipes_2)
recipes_2_scaled =scaler.transform(recipes_2)
Recipes = "scaler.pickle"

with open(Recipes, "wb") as file:
        pickle.dump(scaler,file)
        
recipes_2_scaled_df = pd.DataFrame(recipes_2_scaled, columns = recipes_2.columns)
display(recipes_2.head())

display(recipes_2_scaled_df.head())
```

```python
recipes_2_scaled_df.describe()
```

```python
K = range(10, 25)

inertia = []
silhouette = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234,
                    verbose=1)
    kmeans.fit(recipes_2_scaled_df)
    
    filename = "kmeans_" + str(k) + ".pickle" 
    with open(filename, "wb") as file:
        pickle.dump(kmeans,file)

    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(recipes_2_scaled_df, kmeans.predict(recipes_2_scaled_df)))


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,2,figsize=(12,6))
ax[0].plot(K, inertia, 'bx-')
ax[0].set_xlabel('k')
ax[0].set_ylabel('inertia')
ax[0].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[0].set_title('Elbow Method showing the optimal k')
ax[1].plot(K, silhouette, 'bx-')
ax[1].set_xlabel('k')
ax[1].set_ylabel('silhouette score')
ax[1].set_xticks(np.arange(min(K), max(K)+1, 1.0))
ax[1].set_title('Silhouette Method showing the optimal k')
```

Gaussian Mixtures method

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import pandas as pd

# Select the features you want to use for clustering
selected_features = ['fat', 'calories', 'protein', 'rating', 'sodium', 'len_directions', 'len_ingredients']
X = recipes_2[selected_features].values

# Determine the number of components (clusters) you want to fit
n_components = 16  # Adjust this as needed

# Create a Gaussian Mixture Model
gmm = GaussianMixture(n_components=n_components, random_state=42)

# Fit the GMM to the data
gmm.fit(X)

# Predict cluster assignments for each data point
labels = gmm.predict(X)

# Access the estimated means and covariances of the components
means = gmm.means_
covariances = gmm.covariances_

# Visualize the data points colored by their assigned cluster (for 2D features)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(means[:, 0], means[:, 1], marker='x', s=100, c='red', label='Cluster Centers')
plt.legend()
plt.xlabel(selected_features[0])
plt.ylabel(selected_features[1])
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```

```python
cluster_summary = []

for cluster_id in range(n_components):
    cluster_mask = labels == cluster_id
    cluster_data = X[cluster_mask]
    
    cluster_stats = {
        'Cluster ID': cluster_id,
        'Size': cluster_data.shape[0],
        'Mean': np.mean(cluster_data, axis=0),
        'Median': np.median(cluster_data, axis=0),
        'Std Deviation': np.std(cluster_data, axis=0)
    }
    
    cluster_summary.append(cluster_stats)

cluster_summary_df = pd.DataFrame(cluster_summary)
display(cluster_summary_df)
```

```python
from sklearn.mixture import GaussianMixture

# Predict cluster assignments for the original data
original_data_clusters = gmm.predict(X)

# Assign cluster labels to the DataFrame
recipes_2['cluster'] = original_data_clusters

# Print the count of data points in each cluster
cluster_counts = recipes_2['cluster'].value_counts().sort_index()
display(cluster_counts)

# Display the DataFrame with added cluster labels
display(recipes_2.head())
```
